|转化器            |用例                                  |优点             |缺点           |
|:---|:---|:---|:---|
|html.parser      |BeautifulSoup(markup, "html.parser") |<ul><li>python库自带</li><li>中规中矩的速度</li><li>可应用范围广</li></ul> |<ul><li>速度不如lxml</li><li>应用范围不如html5lib</li></ul> |
|lxml的html转化器   |BeautifulSoup(markup, "lxml")        |<ul><li>速度非常快</li><li>可应用范围广</li></ul>   |<ul><li>外部C语言依赖</li></ul>|
|lxml的xml转化器    |BeautifulSoup(markup, "xml")         |<ul><li>速度非常快</li><li>当前唯一支持XML的转化器</li></ul>|<ul><li>外部C语言依赖</li></ul>|
|html5lib          |BeautifulSoup(markup, "html5lib")    |<ul><li>可应用范围非常广</li><li>类似于浏览器转化过程</li><li>创建有效html5页面</li></ul>|<ul><li>非常慢</li><li>外部python依赖</li></ul>|

BeautifulSoup4对象种类归纳:  
soup = BeautifulSoup('<b class="boldest">Extremely bold</b>')
1. bs4.BeautifulSoup  
    特别的Tag  

2. bs4.element.Tag  
    HTML/XML的tag. 如soup.b  
    tag["id"]/tag["class"]获取属性值  

3. bs4.element.NavigableString  
    tag的正文字符. 如soup.b.string  

4. bs4.element.Comment  
    特别的NavigableString. 格式: \<!-- content -->


继承关系:  
bs4.element.PageElement --> bs4.element.Tag --> bs4.BeautifulSoup  
bs4.element.PageElement, str --> bs4.element.NavigableString --> bs4.element.PreformattedString --> bs4.element.Comment

```
1.bs4.BeautifulSoup

__init__(markup='', features=None, builder=None, parse_only=None, from_encoding=None, exclude_encodings=None, element_classes=None, **kwargs)
    构造BeautifulSoup对象, HTML或XML的数据结构对象. 参数列表如下:
        markup - str格式或bytes格式或file-object, 需要转化的内容
        features - str格式, 指定解析器类型. 列表如下: html.parser/lxml/xml/html5lib. 建议手动指定
                builder - TreeBuilder子类, 不依赖于features构建方法
        parse_only - SoupStrainer对象, 只考虑符合SoupStrainer的文档部分
        from_encoding - str格式, 当markup为bytes格式内容时, 指定markup的编码类型
, 用于解码
        exclude_encodings - list of str格式, 当markup为bytes格式, 排除的编码类型
列表
                element_classes - dict格式, 映射元素的对象类型, 如Tag/NavigableString等
        kwargs - 用于与Beautiful Soup 3后向兼容
```

````python
from bs4 import BeautifulSoup

html_doc = """
<p class="story">
Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a>
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>
</p>
<p class="hungry">
<a href="http://www.baidu.com">catch</a>
<a href="http://www.google.com">try</a>
</p>
"""
soup = BeautifulSoup(html_doc)
list = soup.a
print(list)
````
