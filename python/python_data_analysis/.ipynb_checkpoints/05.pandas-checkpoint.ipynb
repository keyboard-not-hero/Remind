```python
# 例1.测试Series是否包含指定索引值
import pandas as pd

obj2 = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])
print('b' in obj2)
```

```python
# 例2. Series values的花式索引和元素级运算
import pandas as pd

obj2 = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])
print(obj2[obj2 > 0])
print(obj2 * 2)
```

```python
# 例3. DataFrame
import pandas as pd

# dict的key作为DataFrame的列明. DataFrame的每一列都是Series
data = {
        'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]
}
df = pd.DataFrame(data)
print(df)
# 取列与行内容, df['year']也可以表示为df.year
print(df['year']) 
print(df.loc[3])
# 取头几行内容
print(df.head())
```

```python
# 例4.DataFrame与Series进行计算
import numpy as np
import pandas as pd

frame = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns=list('bde'), index=['Utah', 'Ohio', 'Texas', 'Oregon'])
series = frame.iloc[0]
print(frame.sub(series, axis='columns'))
series2 = frame['b']
print(frame.sub(series2, axis='index'))
```
