```python
# 例1.初次使用matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# subplots生成一个figure, 包含一个axis
fig, axis = plt.subplots()
axis.plot(x, y)
plt.show()
```

```python
# 例2.手动创建figure和多个axes
import matplotlib.pyplot as plt

figure, axes = plt.subplots(1, 2)
axes[0].plot([1, 2, 3, 4], [1, 4, 2, 3])
axes[1].plot([1, 2, 3, 4], [3, 2, 4, 1])
plt.show()
```

```python
# 例3.interpolation类型
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(120).reshape((10, 12))

interp1 = 'nearest'
interp2 = 'bilinear'
interp3 = 'bicubic'
fig, axs = plt.subplots(ncols=3, sharex=True, figsize=(15, 5))
axs[0].set_title('nearest')
axs[0].imshow(x, origin='upper', interpolation=interp1)

axs[1].set_title('bilinear')
axs[1].imshow(x, origin='upper', interpolation=interp2)

axs[2].set_title('bicubic')
axs[2].imshow(x, origin='upper', interpolation=interp3)
plt.show()
```

```python
# 例4. 配置参数(当前session持续)
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np


# matplotlib模块
# matplotlib.rcParams
#     参考列表: https://matplotlib.org/stable/api/matplotlib_configuration_api.html#matplotlib.rcParams
mpl.rcParams['lines.linewidth'] = 1
mpl.rcParams['lines.linestyle'] = '--'
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)
fig, axes = plt.subplots()
axes.plot(x, y)
plt.show()
```

```python
# 例5.设置字体[1]
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import numpy as np


x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y = np.sin(x)
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.plot(x, y, scalex=True, scaley=True, linewidth=2, label='line1')
axes.set_title("sin(x)", loc='right', y=1)
prop = fm.FontProperties(fname="/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc")
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = prop.get_name()
axes.set_xlabel("x轴", labelpad=4.0, loc='center')
plt.show()
```
